# create database connection

```python
# When you run this line, SQLAlchemy doesn't actually connect to the database yet. Instead, it creates a "factory" (the engine object) that is ready to open a connection the moment you try to perform a task, like querying a table or inserting data.

# Why use create_engine?
# Instead of writing raw, complex connection code every time, the engine handles:

# Connection Pooling: Keeping a few connections open so you don't have to "log in" to the DB for every single tiny query.

# Dialect Handling: It translates Python commands into the specific flavor of SQL that PostgreSQL understands.

from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg://root:root@localhost:5432/ny_taxi')
```

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```
